```python
from __future__ import division
import numpy as np
import sys
```

```python
def derivs(e):
    
    de = e*(1-e**2)   
    
    return de
```

```python
t = 0
nsteps = 100
h = 0.1
Tf = nsteps*h
counter = 0
eorig = 0.5
#nsteps = int(Tf / h) 
out = np.zeros((nsteps,2)) #t,e
yn = eorig
tt = []
yy = []
while t < Tf:

    k1 = h * derivs(yn)
    k2 = h * derivs(yn + k1/2)
    k3 = h * derivs(yn + k2/2)
    k4 = h * derivs(yn + k3)
    ynew = yn + (k1 + 2*k2 + 2*k3 + k4)/6

    
    yn = ynew
   
    


    
    t = t + h
    
    tt.extend([t])
    yy.extend([yn])
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt


fig = plt.figure(figsize=(10,10))
ax1 = plt.subplot2grid((1,1), (0,0)) 
ax1.plot(tt,yy)
```
